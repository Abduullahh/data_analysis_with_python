# Model development

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv')
```

```python
df.head()
```

```python
lm = LinearRegression()
```

## SLR

```python
x = df[['highway-mpg']]
y = df[['price']]
```

```python
lm.fit(x, y)
```

```python
yhat = lm.predict(x)
```

```python
lm.intercept_
```

```python
lm.coef_
```

### The relationship is :
**<code>price = 38423.31 - 821.73*highway-mpg</code>**

## MLR

```python
z = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
```

```python
z.head()
```

```python
lm.fit(z, df['price'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

## The relationship is :
**<code>price = -15806.62 + (53.5*horsepower) + (4.7*curb-weight) + (81.5*engine-size) + (36.1*highway-mpg)</code>**
